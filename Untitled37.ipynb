```python
from google.colab import drive
drive.mount('/content/drive')
!pip install google-play-scraper
import os
import time
import pandas as pd
from google_play_scraper import reviews, Sort
# Set your Google Drive save path
save_path = "/content/drive/MyDrive/Health_Reviews"
os.makedirs(save_path, exist_ok=True)

# List of Health & Wellness apps
health_apps = [
    {"name": "Medisafe_Pill_Reminder", "package": "com.medisafe.android.client"},
    {"name": "MyFitnessPal", "package": "com.myfitnesspal.android"},
    {"name": "Blood_Pressure_Diary", "package": "com.frmdstudio.bloodpressurediary"},
    {"name": "Samsung_Health", "package": "com.sec.android.app.shealth"},
    {"name": "Calm", "package": "com.calm.android"},
    {"name": "HealthTap", "package": "com.healthtap.userhtexpress"},
    {"name": "Google_Fit", "package": "com.google.android.apps.fitness"},
    {"name": "CareClinic", "package": "com.careclinic.careclinicapp"},
    {"name": "Fitbit", "package": "com.fitbit.FitbitMobile"},
    {"name": "Ada_Health_Check", "package": "com.ada.app"}
]
def collect_next_chunk(app, chunk_size=30000):
    all_reviews = []
    next_token = None
    collected = 0

    print(f"\nğŸ”„ Collecting next {chunk_size} reviews for: {app['name']}")
    while collected < chunk_size:
        try:
            result, next_token = reviews(
                app['package'],
                lang='en',
                country='us',
                sort=Sort.NEWEST,
                count=200,
                continuation_token=next_token
            )

            if not result:
                print("âš ï¸ No more reviews available.")
                break

            all_reviews.extend(result)
            collected += len(result)

            if collected % 1000 == 0:
                print(f"âœ… Total collected: {collected}")

            if not next_token:
                break

            time.sleep(1)
        except Exception as e:
            print(f"âŒ Error: {e}")
            break

    # Save to CSV
    if all_reviews:
        df = pd.DataFrame(all_reviews)
        df['app_name'] = app['name']
        df['category'] = "Health & Wellness"

        timestamp = int(time.time())
        filename = f"{app['name']}_chunk_{timestamp}.csv"
        full_path = os.path.join(save_path, filename)

        df[['reviewId', 'userName', 'score', 'content', 'at', 'app_name', 'category']].to_csv(full_path, index=False)
        print(f"ğŸ“ Saved: {full_path}")
    else:
        print("âš ï¸ No reviews collected in this chunk.")
        collect_next_chunk(health_apps[0], chunk_size=30000)
collect_next_chunk(health_apps[1], chunk_size=30000)
import glob

# List all chunk files for an app
csv_files = glob.glob("/content/drive/MyDrive/Health_Reviews/Medisafe_Pill_Reminder_chunk_*.csv")

# Merge them
df_list = [pd.read_csv(f) for f in csv_files]
merged_df = pd.concat(df_list, ignore_index=True)

# Save merged file
merged_df.to_csv("/content/drive/MyDrive/Health_Reviews/Medisafe_Pill_Reminder_merged.csv", index=False)
```
