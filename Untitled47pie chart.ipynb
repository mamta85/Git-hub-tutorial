```python

from google.colab import drive
drive.mount('/content/drive')
# /content/drive/MyDrive/Health_Reviews/Medisafe_Pill_Reminder_chunk_1749543678.csv
import pandas as pd

file_path = '/content/drive/MyDrive/Health_Reviews/Medisafe_Pill_Reminder_chunk_1749543678.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Display first few rows to confirm
df.head()
print(df.columns)
df['content'] = df['content'].astype(str).str.lower()
# Define keywords
complaint_keywords = ['issue', 'bug', 'error', 'problem', 'slow', 'crash', 'freeze',
                      'not working', 'ads', 'glitch', 'fail', "doesn't", "can't",
                      'bad', 'hang', 'stuck', 'uninstall', 'broken', 'fix', 'lag']

suggestion_keywords = ['add', 'include', 'suggest', 'should', 'please', 'feature',
                       'recommend', 'option', 'would be better', 'could', 'need', 'wish',
                       'want', 'request', 'provide', 'improve', 'enhance', 'make it',
                       'allow', 'support']
                       # Define functions
def contains_keywords(text, keywords):
    return any(keyword in text for keyword in keywords)

# Apply functions
df['is_complaint'] = df['content'].apply(lambda x: contains_keywords(x, complaint_keywords))
df['is_suggestion'] = df['content'].apply(lambda x: contains_keywords(x, suggestion_keywords))
num_complaints = df['is_complaint'].sum()
num_suggestions = df['is_suggestion'].sum()

print(f"Total Complaints: {num_complaints}")
print(f"Total Suggestions: {num_suggestions}")
import matplotlib.pyplot as plt

labels = ['Complaints', 'Suggestions']
sizes = [num_complaints, num_suggestions]
colors = ['red', 'green']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('User Feedback Distribution')
plt.axis('equal')
plt.show()
print(df.columns)











```
