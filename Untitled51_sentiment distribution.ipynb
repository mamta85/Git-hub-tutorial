```python
# Apply sentiment
# Ensure the 'content' column is of string type before applying the function
df['sentiment'] = df['content'].astype(str).apply(get_sentiment)

# Suggestion and complaint keyword lists
suggestion_keywords = ['add', 'include', 'suggest', 'should', 'please', 'feature', 'recommend', 'option', 'would be better', 'could', 'need', 'wish', 'want', 'request', 'provide', 'improve', 'enhance', 'make it', 'allow', 'support']
complaint_keywords = ['issue', 'bug', 'error', 'problem', 'slow', 'crash', 'freeze', 'not working', 'ads', 'glitch', 'fail', "doesn't", "can't", 'bad', 'hang', 'stuck', 'uninstall', 'broken', 'fix', 'lag']

# Subsets
df_suggestions = df[df['content'].str.contains('|'.join(suggestion_keywords), case=False, na=False)]
df_complaints = df[df['content'].str.contains('|'.join(complaint_keywords), case=False, na=False)]
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Suggestions
sns.countplot(x='sentiment', data=df_suggestions, ax=axes[0], palette='Greens')
axes[0].set_title("✅ Sentiment Distribution in Suggestion Reviews")
axes[0].set_xlabel("Sentiment")
axes[0].set_ylabel("Count")

# Complaints
sns.countplot(x='sentiment', data=df_complaints, ax=axes[1], palette='Reds')
axes[1].set_title("❌ Sentiment Distribution in Complaint Reviews")
axes[1].set_xlabel("Sentiment")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()




```
