```python


# Step 1: Upload your file (Excel or CSV)
from google.colab import files
uploaded = files.upload()

# Step 2: Import required libraries
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Step 3: Detect and load file
filename = list(uploaded.keys())[0]
print(f"Uploaded file: {filename}")

if filename.endswith('.csv'):
    df = pd.read_csv(filename)
    print("✅ Loaded CSV file successfully!")
elif filename.endswith('.xlsx') or filename.endswith('.xls'):
    try:
        df = pd.read_excel(filename, engine='openpyxl')
        print("✅ Loaded Excel file successfully!")
    except Exception as e:
        print(f"❌ Failed to load Excel file: {e}")
else:
    raise Exception("Unsupported file format. Please upload a CSV or Excel file.")

# Step 4: Show column names and strip whitespace
print("\nColumns in your file:")
df.columns = df.columns.str.strip()
print(df.columns.tolist())

# Step 5: Use the correct column that contains reviews
review_column_name = 'text_en'  # ✅ Use 'content' if you prefer the original language

# Step 6: Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
# Download the specific punkt_tab resource required by word_tokenize
nltk.download('punkt_tab')

# Step 7: Define preprocessing function
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = word_tokenize(str(text).lower())
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return ' '.join(tokens)

# Step 8: Apply preprocessing to review column
df['cleaned_review'] = df[review_column_name].apply(preprocess)

# Step 9: Create TF-IDF matrix
vectorizer = TfidfVectorizer(max_df=0.8, min_df=2, stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['cleaned_review'])
feature_names = vectorizer.get_feature_names_out()

# Step 10: Print top keywords from first review
first_vector = tfidf_matrix[0]
sorted_indices = np.argsort(first_vector.toarray()).flatten()[::-1]

print("\nTop TF-IDF keywords in first review:")
for idx in sorted_indices[:10]:
    print(f"{feature_names[idx]}: {first_vector[0, idx]:.4f}")
```
