```python
import matplotlib.pyplot as plt

# Convert counters to sorted lists
sorted_suggestions = sorted(suggestion_counts.items(), key=lambda x: x[1], reverse=True)
sorted_complaints = sorted(complaint_counts.items(), key=lambda x: x[1], reverse=True)

# Separate keywords and counts
sugg_keywords, sugg_values = zip(*sorted_suggestions)
comp_keywords, comp_values = zip(*sorted_complaints)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Suggestion Keyword Bar Plot
axes[0].barh(sugg_keywords[::-1], sugg_values[::-1], color='green')
axes[0].set_title('✅ Suggestion Keywords Frequency')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Keyword')

# Complaint Keyword Bar Plot
axes[1].barh(comp_keywords[::-1], comp_values[::-1], color='red')
axes[1].set_title('❌ Complaint Keywords Frequency')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Keyword')

plt.tight_layout()
plt.show()




```
