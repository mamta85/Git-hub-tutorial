```python
import pandas as pd

file_path = '/content/drive/MyDrive/Health_Reviews/Combined_Cleaned_Reviews.csv'

try:
    df = pd.read_csv(file_path)
    print("✅ File loaded successfully. Preview:")
    print(df.head())
except FileNotFoundError:
    print("❌ Error: File not found at", file_path)
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# Replace path with your correct file path
file_path = '/content/drive/MyDrive/Health_Reviews/Combined_Cleaned_Reviews.csv'
df = pd.read_csv(file_path)

# Show first few rows
df.head()
issue_keywords = ['issue', 'bug', 'error', 'problem', 'slow', 'crash', 'freeze',
                  'not working', 'ads', 'glitch', 'fail', "doesn't", "can't",
                  'bad', 'hang', 'stuck', 'uninstall', 'broken', 'fix', 'lag']
suggestion_keywords = ['add', 'include', 'suggest', 'should', 'please', 'feature', 'recommend',
                       'option', 'would be better', 'could', 'need', 'wish', 'want',
                       'request', 'provide', 'improve', 'enhance', 'make it', 'allow', 'support']
all_keywords = list(set(issue_keywords + suggestion_keywords))

# Ensure review column exists
df['content'] = df['content'].astype(str).str.lower()
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(vocabulary=all_keywords)
X_tfidf = vectorizer.fit_transform(df['content'])
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)  # You can adjust the number of clusters
df['cluster'] = kmeans.fit_predict(X_tfidf)
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_tfidf.toarray())

plt.figure(figsize=(12, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['cluster'], palette="Set2", legend='full')
plt.title("Clusters of Users Based on Review Issues and Suggestions")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.tight_layout()
plt.show()
df[['user_id', 'content', 'cluster']].to_csv('/content/drive/MyDrive/Health_Reviews/UserClusters.csv', index=False)



```
