```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

filename = list(uploaded.keys())[0]

# If CSV file, load with pandas
df = pd.read_csv(filename)

# Display first few rows to check
df.head()
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    return ' '.join(tokens)

df['processed_text'] = df['text_en'].apply(preprocess_text)  # Change 'text_en' to your review column name
df.head()
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine userId and processed_text for grouping
user_reviews = df.groupby('userName')['processed_text'].apply(lambda texts: ' '.join(texts)).reset_index()

vectorizer = TfidfVectorizer(max_features=1000)  # limit features to top 1000 terms

tfidf_matrix = vectorizer.fit_transform(user_reviews['processed_text'])
feature_names = vectorizer.get_feature_names_out()

# Create DataFrame: rows=userName, columns=issues/features, values=TF-IDF scores
user_issue_matrix = pd.DataFrame(tfidf_matrix.toarray(), index=user_reviews['userName'], columns=feature_names)

user_issue_matrix.head()
import matplotlib.pyplot as plt
import seaborn as sns

top_n = 20
top_features = user_issue_matrix.mean(axis=0).sort_values(ascending=False).head(top_n).index

plt.figure(figsize=(12, 8))
sns.heatmap(user_issue_matrix[top_features], cmap="YlGnBu", linewidths=0.5)
plt.title(f"Heatmap of Top {top_n} Issues Across Users")
plt.xlabel("Issues (Features)")
plt.ylabel("Users")
plt.show()
from sklearn.cluster import KMeans

X = user_issue_matrix[top_features].fillna(0)
num_clusters = 3

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
user_clusters = kmeans.fit_predict(X)

user_issue_matrix['cluster'] = user_clusters
print(user_issue_matrix['cluster'].value_counts())
user_issue_matrix[['cluster']].head(10)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10,7))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=user_clusters, cmap='Set1')
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.title("User Clusters based on App Issues (PCA projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
