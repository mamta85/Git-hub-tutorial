```python
# Step 1: Upload CSV
from google.colab import files
uploaded = files.upload()

# Step 2: Import Libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Step 3: Load Data
df = pd.read_csv("Reviews_with_TFIDF.csv")
print("Shape of dataset:", df.shape)
display(df.head()) # Use display for better formatting in notebooks
# Step 4: Inspect columns and separate TF-IDF features
print("Columns:", df.columns.tolist())

# Replace 'review_text' with the actual review column name in your file
# Identify the non-numerical columns that are not TF-IDF features
# Correctly identify and exclude the review text column ('content') and other non-TF-IDF columns
# Assuming 'content' is the review text column containing strings
non_feature_cols = ['reviewId', 'userName', 'score', 'content'] # Add 'content' to columns to exclude

# Select only the TF-IDF feature columns (numeric columns that are not the excluded non-feature columns)
# We assume TF-IDF features are all numeric. We filter by dtype to ensure only numeric columns are selected.
tfidf_features = df.select_dtypes(include=np.number).drop(columns=[col for col in non_feature_cols if col in df.columns], errors='ignore')

# Store the name of the actual text column for later use
text_col = 'content' # Set the name of the text column

# Step 5: Dimensionality Reduction for Visualization
pca = PCA(n_components=2)
# Ensure tfidf_features only contains numerical data before fitting PCA
reduced_data = pca.fit_transform(tfidf_features)

# Step 6: Apply KMeans Clustering
k = 5  # You can adjust this number based on dataset size
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Add n_init explicitly to suppress warning
df['cluster'] = kmeans.fit_predict(tfidf_features)

# Step 7: Visualize Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=reduced_data[:, 0], y=reduced_data[:, 1], hue=df['cluster'], palette='Set1')
plt.title("User Communities Based on Similar Reviews")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title='Cluster')
plt.show()

# Step 8: Sample Reviews from Each Cluster
for i in range(k):
    print(f"\nCluster {i} - Sample Reviews:")
    # Make sure to handle potential empty clusters
    if not df[df['cluster'] == i].empty:
        print(df[df['cluster'] == i][text_col].sample(5, random_state=42).values)
    else:
        print("No samples in this cluster.")


# Step 9: Complaint/Suggestion Tagging
complaint_keywords = ['error', 'crash', 'bug', 'slow', 'not working', 'issue', 'problem', 'lag', 'freeze']
suggestion_keywords = ['please add', 'it would be better', 'suggest', 'recommend', 'should have', 'feature request']

def tag_review(text):
    text = str(text).lower()
    if any(word in text for word in complaint_keywords):
        return 'complaint'
    elif any(word in text for word in suggestion_keywords):
        return 'suggestion'
    else:
        return 'other'

df['tag'] = df[text_col].apply(tag_review)

# Step 10: Cluster-wise Tag Distribution
cluster_tag_dist = df.groupby(['cluster', 'tag']).size().unstack(fill_value=0)
print("\nCluster-wise Complaint/Suggestion Distribution:")
print(cluster_tag_dist)

# Optional: Visualize Tag Distribution
cluster_tag_dist.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title("Cluster-wise Distribution of Complaints and Suggestions")
plt.xlabel("Cluster")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=0)
plt.legend(title="Tag")
plt.show()







```
