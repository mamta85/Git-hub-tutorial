```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import os

# Path to your cleaned review file
folder_path = '/content/drive/MyDrive/Health_Reviews'
file_path = os.path.join(folder_path, 'Combined_Cleaned_Reviews.csv')

# Load data
df = pd.read_csv(file_path)
print("‚úÖ Data loaded:", df.shape)
df[['cleaned_content']].head()
# Install if needed
!pip install scikit-learn

# Import TF-IDF tools
from sklearn.feature_extraction.text import TfidfVectorizer
# Create TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, min_df=5, max_features=1000)

# Fit and transform on cleaned_content
tfidf_matrix = tfidf_vectorizer.fit_transform(df['cleaned_content'].astype(str))

# Get feature names (i.e., important words)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Convert to DataFrame
import numpy as np
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# Show TF-IDF scores (mean across documents)
mean_tfidf = tfidf_df.mean().sort_values(ascending=False)
mean_tfidf.head(20)
# Top 30 keywords indicating most common issues
top_issues = mean_tfidf.head(30)
top_issues_df = top_issues.reset_index()
top_issues_df.columns = ['Keyword', 'TF-IDF Score']
print("üîù Top 30 Keywords (Issues):")
print(top_issues_df)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
top_issues.plot(kind='bar', color='skyblue')
plt.title('Top 30 Keywords (User Issues via TF-IDF)')
plt.xlabel('Keyword')
plt.ylabel('Average TF-IDF Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
output_issues_path = os.path.join(folder_path, "Top_30_Issues_TFIDF.csv")
top_issues_df.to_csv(output_issues_path, index=False)
print(f"‚úÖ Top issues saved to: {output_issues_path}")
```
