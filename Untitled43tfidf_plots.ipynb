```python


# Step 1: Import required libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 2: Mount Google Drive (if not already mounted)
from google.colab import drive
drive.mount('/content/drive')

# Step 3: Load the reviews file
file_path = "/content/drive/MyDrive/Health_Reviews/Combined_Cleaned_Reviews.csv"  # âœ… Update path
df = pd.read_csv(file_path)

# Step 4: Define the issue-related keywords
issue_keywords = [
    'issue', 'bug', 'error', 'problem', 'slow', 'crash', 'freeze',
    'not working', 'ads', 'glitch', 'fail', "doesn't", "can't",
    'bad', 'hang', 'stuck', 'uninstall', 'broken', 'fix', 'lag'
]
suggestion_keywords = [
    'add', 'include', 'suggest', 'should', 'please', 'feature', 'recommend',
    'option', 'would be better', 'could', 'need', 'wish', 'want',
    'request', 'provide', 'improve', 'enhance', 'make it', 'allow', 'support'
]

# Step 5: Preprocess text data
corpus = df['content'].astype(str).tolist()  # âœ… Update 'content' to your review column if needed

# Step 6: Apply TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()
tfidf_scores = X.mean(axis=0).A1  # Get average TF-IDF scores

# Step 7: Create TF-IDF DataFrame
tfidf_df = pd.DataFrame({
    'Keyword': feature_names,
    'TF-IDF Score': tfidf_scores
})

# Step 8: Filter for issue-related keywords
# Renamed filtered_df to issue_df for clarity in plotting later
issue_df = tfidf_df[tfidf_df['Keyword'].isin([kw.lower() for kw in issue_keywords])]
issue_df = issue_df.sort_values(by='TF-IDF Score', ascending=False).reset_index(drop=True)

# Step 8a: Filter for suggestion-related keywords (Added)
suggestion_df = tfidf_df[tfidf_df['Keyword'].isin([kw.lower() for kw in suggestion_keywords])]
suggestion_df = suggestion_df.sort_values(by='TF-IDF Score', ascending=False).reset_index(drop=True)


# Step 9: Display the filtered result
print("ðŸ“Š TF-IDF Scores for Issue-Related Keywords:")
print(issue_df.to_string(index=True))
print("\nðŸ’¡ TF-IDF Scores for Suggestion-Related Keywords:") # Added print for suggestion keywords
print(suggestion_df.to_string(index=True))


# Step 10 (Optional): Save output to CSV
issue_df.to_csv("/content/drive/MyDrive/Health_Reviews/TFIDF_Issue_Keywords.csv", index=False)
suggestion_df.to_csv("/content/drive/MyDrive/Health_Reviews/TFIDF_Suggestion_Keywords.csv", index=False) # Added saving for suggestion keywords

import matplotlib.pyplot as plt
import seaborn as sns
# Set up the figure for issue keywords
plt.figure(figsize=(10, 6))
sns.barplot(x='TF-IDF Score', y='Keyword', data=issue_df, palette='Reds_r')
plt.title('ðŸ”¥ Top TF-IDF Scores for Issue Keywords', fontsize=14)
plt.xlabel('TF-IDF Score')
plt.ylabel('Keyword')
plt.tight_layout()
plt.show()

# Set up the figure for suggestion keywords
plt.figure(figsize=(10, 6))
sns.barplot(x='TF-IDF Score', y='Keyword', data=suggestion_df, palette='Blues_d')
plt.title('ðŸ’¡ Top TF-IDF Scores for Suggestion Keywords', fontsize=14)
plt.xlabel('TF-IDF Score')
plt.ylabel('Keyword')
plt.tight_layout()
plt.show()

# Save plots to file
plt.figure(figsize=(10, 6))
sns.barplot(x='TF-IDF Score', y='Keyword', data=issue_df, palette='Reds_r')
plt.title('Top TF-IDF Scores for Issue Keywords')
plt.savefig("/content/drive/MyDrive/YourFolder/Issue_TFIDF_Barchart.png", bbox_inches='tight')
plt.close()

plt.figure(figsize=(10, 6))
sns.barplot(x='TF-IDF Score', y='Keyword', data=suggestion_df, palette='Blues_d')
plt.title('Top TF-IDF Scores for Suggestion Keywords')
plt.savefig("/content/drive/MyDrive/YourFolder/Suggestion_TFIDF_Barchart.png", bbox_inches='tight')
plt.close()
```
