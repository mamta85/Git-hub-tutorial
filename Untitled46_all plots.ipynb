```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (update the filename if different)
file_path = '/content/drive/MyDrive/Health_Reviews/Medisafe_Pill_Reminder_chunk_1749543678.csv'
df = pd.read_csv(file_path)

# Convert review column to lowercase for case-insensitive matching
df['content'] = df['content'].astype(str).str.lower()

# Define complaint and suggestion keywords
complaint_keywords = ['issue', 'bug', 'error', 'problem', 'slow', 'crash', 'freeze',
                      'not working', 'ads', 'glitch', 'fail', "doesn't", "can't",
                      'bad', 'hang', 'stuck', 'uninstall', 'broken', 'fix', 'lag']

suggestion_keywords = ['add', 'include', 'suggest', 'should', 'please', 'feature', 'recommend',
                       'option', 'would be better', 'could', 'need', 'wish', 'want',
                       'request', 'provide', 'improve', 'enhance', 'make it', 'allow', 'support']

# Create flags for complaints and suggestions
df['is_complaint'] = df['content'].apply(lambda x: any(keyword in x for keyword in complaint_keywords))
df['is_suggestion'] = df['content'].apply(lambda x: any(keyword in x for keyword in suggestion_keywords))

# Find reviews that are both complaints and suggestions
df['both_complaint_suggestion'] = df['is_complaint'] & df['is_suggestion']

# Count categories
total_reviews = len(df)
only_complaints = df[(df['is_complaint']) & (~df['is_suggestion'])].shape[0]
only_suggestions = df[(~df['is_complaint']) & (df['is_suggestion'])].shape[0]
both = df['both_complaint_suggestion'].sum()
neither = df[(~df['is_complaint']) & (~df['is_suggestion'])].shape[0]

# Summary table
summary = {
    'Only Complaints': only_complaints,
    'Only Suggestions': only_suggestions,
    'Both Complaints & Suggestions': both,
    'Neither': neither,
    'Total Reviews': total_reviews
}

summary_df = pd.DataFrame(summary.items(), columns=['Category', 'Count'])
print(summary_df)

# Show 5 examples of reviews that are both complaint and suggestion
print("\nüîç Sample reviews (both complaint and suggestion):")
print(df[df['both_complaint_suggestion']]['content'].head())

# Pie chart
labels = ['Only Complaints', 'Only Suggestions', 'Both', 'Neither']
sizes = [only_complaints, only_suggestions, both, neither]
colors = ['red', 'green', 'orange', 'gray']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('üìä Review Classification: Complaints vs Suggestions')
plt.axis('equal')
plt.show()


```
