```python
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Health_Reviews/TFIDF_Keywords_Result.csv'
import pandas as pd

# Load the CSV
df = pd.read_csv(file_path)

# See all column names
print(df.columns)
# Use correct column name instead of 'review'
# Replace 'YOUR_CORRECT_COLUMN_NAME' with the actual column name from the print(df.columns) output
review_col = 'content'

# Define keywords
issue_keywords = ['issue', 'bug', 'error', 'problem', 'slow', 'crash', 'freeze']
suggestion_keywords = ['suggest', 'recommend', 'should', 'need', 'wish', 'prefer']

# Mark rows with issues or suggestions
df['has_issue'] = df[review_col].apply(lambda x: any(k in str(x).lower() for k in issue_keywords))
df['has_suggestion'] = df[review_col].apply(lambda x: any(k in str(x).lower() for k in suggestion_keywords))
incidence_matrix = df[['content', 'has_issue', 'has_suggestion']]
# Corrected the directory name, removing the extra space
incidence_matrix.to_csv('/content/drive/MyDrive/Health_Reviews/TFIDF_Keywords_Result.csv', index=False)
incidence_matrix.head(10)
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = incidence_matrix[['has_issue', 'has_suggestion']]
pca = PCA(n_components=2)
components = pca.fit_transform(X)

plt.figure(figsize=(8, 5))
plt.scatter(components[:, 0], components[:, 1], c='blue')
plt.title('User Clusters Based on Feedback')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()

```
